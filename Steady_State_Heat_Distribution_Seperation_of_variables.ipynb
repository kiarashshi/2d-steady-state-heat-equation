```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad


L = 1.0
W = 1.0


N = 30


def f(x):
    return 1  # u(x, W) = 1

x = np.linspace(0, L, 100)
y = np.linspace(0, W, 100)
X, Y = np.meshgrid(x, y)


u = np.zeros_like(X)


for n in range(1, N+1):
    lambda_n = (n * np.pi / L) ** 2
    A_n, _ = quad(lambda x: f(x) * np.sin(n * np.pi * x / L), 0, L)
    A_n *= 2 / (L * np.sinh(n * np.pi * W / L))
    u += A_n * np.sin(n * np.pi * X / L) * np.sinh(n * np.pi * Y / L)


plt.contourf(X, Y, u, 50, cmap='hot')
plt.colorbar(label='Temperature')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Steady-State Heat Distribution-Seperation-of-variables')
plt.show()
```
